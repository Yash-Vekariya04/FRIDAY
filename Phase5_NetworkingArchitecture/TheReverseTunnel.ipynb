## 3-Steps Implementation

### Step-1: Start F.R.I.D.A.Y. locally
You write your FastAPI Python script and run it on the Pi. It sits there listening on port 8000.

    `uvicorn friday_server:app --host 127.0.0.1 --port 8000`

### Step 2: Install the Tunnel
You open a second terminal window on the Pi and download the cloudflared daemon (a tiny background program written in C/Go).

    `sudo apt install cloudflared`

### Step 3: Link the Tunnel to the Brain
You run one final command to tell Cloudflare to map a public URL directly to your local port 8000.

    `cloudflared tunnel --url http://127.0.0.1:8000`

The terminal will instantly spit out a random, secure URL (like https://random-words-123.trycloudflare.com).

You can copy that link, send it to a friend in another country, and if they click it, they will instantly see your custom F.R.I.D.A.Y. web dashboard and be able to talk to the AI running on the little board sitting on your desk in Agartala.
